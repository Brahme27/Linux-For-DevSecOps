## 1 Software Development Life Cycle (SDLC)

### 1.1 Meaning of SDLC

SDLC stands for **Software Development Life Cycle**. It is the complete step-by-step process followed to develop software from the beginning until maintenance. Every software project follows SDLC in some form.

### 1.2 Phases of SDLC

The first phase is **Requirements Analysis**, where we understand what the client needs. After that comes **Planning**, where timelines, cost, and resources are decided. Then comes **Design**, where general requirements are converted into technical requirements. After design, developers start **Implementation**, which means writing code. Once coding is completed, the application goes to the **Testing** phase to find bugs and defects. After testing, the software is **Deployed** into the environment. Finally, **Maintenance** is done to fix issues and add updates after release.

---

## 2 Waterfall Model

### 2.1 What is Waterfall

The Waterfall model is a traditional and linear approach of software development. In this model, work is done phase by phase. Once we move to the next phase, we **cannot go back** and change previous requirements.

---

### 2.2 School Management System (SMS) Example

To understand Waterfall easily, we use the **School Management System** example.

#### Old System (100 Years Back)

Earlier, schools had only one **final exam**. Teachers were not serious about completing the syllabus from day one. Students were also not serious about studying from day one. Parents were only worried whether their child would pass or fail at the end. Because of this approach, only **30 percent seriousness** existed in the system.

---

### 2.3 Process Change in Education

Later, schools introduced **Unit Test I, II, III, IV, Quarterly, Half-Yearly, Pre-Final, and Final exams**. Unit Test I happens after 30 days. Because of this, teachers had to start teaching seriously from day one. Students also started studying earlier. Parents began waiting for unit test results. Due to this change, seriousness increased to **80 percent**.

---

### 2.4 Mapping School Roles to IT

In IT terms, **Parents are Clients**, **Students are Developers, Testing, and Operations teams**, and **Teachers are IT Management**.

---

### 2.5 Waterfall Defects and Efficiency

In Waterfall, development may take **six months**, and testing and deployment happen at the end. In one project, there can be **100 defects**, out of which **10 may be invalid defects**. Sometimes there may be **50 defects with 6 invalid defects**. Waterfall is compared to an **Ambassador car**, which is slow but steady.

---

## 3 Agile Model

### 3.1 What is Agile

Agile is an improved model where the project is divided into small modules instead of doing everything at once.

### 3.2 Agile Modules

Common modules in Agile are **Signup, Login, Menu, Order, Shipping, Delivery, and Payments**.

---

### 3.3 Agile Working Style

Each module is developed separately. For example, the Signup module may take **one month**, where **two weeks are used for development** and **two weeks for testing and deployment**. In Agile, development starts seriously from **day one**.

---

### 3.4 Defects in Agile

Because testing happens continuously, defects are reduced. For example, there may be **20 defects**, out of which only **5 are invalid defects**. Agile is compared to a **Honda City**, which is faster and more efficient than Waterfall.

---

## 4 DevOps

### 4.1 Testing Frequency Comparison

In **Waterfall**, testing happens around **10 times**, and the score is 100.
In **Agile**, testing happens around **30 times**, and the score is 101.
In **DevOps**, testing happens around **100 times**, and the score is 102.

---

### 4.2 Agile as Part of DevOps

Agile is a part of DevOps. DevOps takes Agile practices and improves them further using automation and collaboration.

---

### 4.3 DevOps Working Example

In DevOps, even if a developer writes **100 lines of code on day one**, such as entering first name and last name, that code is immediately **deployed and tested**. If **three defects** are found, teams fix them using **Collaboration, Cooperation, and Coordination**. By **day ten**, everything is already tested and deployed.

---

### 4.4 Definition of DevOps

DevOps is a process of **building, testing, and releasing code on the same day** when the developer writes it. It helps achieve **speed and accuracy**, faster releases, and fewer defects.

---

### 4.5 DevOps Deployment Pipelines

DevOps pipelines evolve step by step:

* DEV to PROD
* DEV to QA to PROD
* DEV to QA to UAT to PRE-PROD to PROD
* Full pipeline with DEV, QA, UAT, PERF, SECURITY, PRE-PROD, and PROD

---

## 5 Linux

### 5.1 Linux vs Windows

Linux is preferred over Windows for servers and DevOps. Windows needs frequent restarts, while Linux can run for years. Windows has heavy graphics and takes time to load, but Linux has no graphics and provides better performance. Windows consumes more resources, while Linux consumes less. Windows is costly and less secure, while Linux is free, secure, and open source.
