# 1. File & Directory Operations

## Basic Navigation & Listing

### `pwd` - Print Working Directory
Shows the current directory path.

```bash
# Syntax
pwd [options]

# Example
pwd
# Output: /home/devops/projects
```

### `ls` - List Directory Contents
Lists files and directories.

```bash
# Syntax
ls [options] [path]

# Examples
ls                  # List files in current directory
ls -l               # Long format with permissions, size, date
ls -la              # Include hidden files (starting with .)
ls -lh              # Human-readable file sizes
ls -lS              # Sort by file size (largest first)
ls -lt              # Sort by modification time (newest first)
ls -lR              # Recursive listing
ls -i               # Show inode numbers
ls *.txt            # List only .txt files

# Output example (ls -la):
# drwxr-xr-x  5 user group 4096 Dec 27 10:00 .
# drwxr-xr-x  3 user group 4096 Dec 26 09:00 ..
# -rw-r--r--  1 user group 1234 Dec 27 10:00 file.txt
```

### `cd` - Change Directory
Navigate between directories.

```bash
# Syntax
cd [directory]

# Examples
cd /var/log         # Go to absolute path
cd ..               # Go up one directory
cd ../..            # Go up two directories
cd ~                # Go to home directory
cd -                # Go to previous directory
cd                  # Go to home directory (same as cd ~)
cd /                # Go to root directory
```

### `mkdir` - Make Directory
Create new directories.

```bash
# Syntax
mkdir [options] directory_name

# Examples
mkdir mydir                     # Create single directory
mkdir dir1 dir2 dir3            # Create multiple directories
mkdir -p /path/to/nested/dir    # Create parent directories if needed
mkdir -m 755 mydir              # Create with specific permissions
mkdir -v mydir                  # Verbose output
```

### `rmdir` - Remove Empty Directory
Remove empty directories only.

```bash
# Syntax
rmdir [options] directory_name

# Examples
rmdir mydir                     # Remove empty directory
rmdir -p path/to/dir            # Remove directory and parents if empty
rmdir -v mydir                  # Verbose output
```

### `touch` - Create Empty File / Update Timestamp
Create files or update timestamps.

```bash
# Syntax
touch [options] filename

# Examples
touch newfile.txt               # Create empty file
touch file1.txt file2.txt       # Create multiple files
touch -a file.txt               # Update access time only
touch -m file.txt               # Update modification time only
touch -d "2025-01-01" file.txt  # Set specific date
touch -r ref.txt file.txt       # Use timestamp from reference file
```

## File Operations

### `cp` - Copy Files and Directories
Copy files and directories.

```bash
# Syntax
cp [options] source destination

# Examples
cp file.txt backup.txt              # Copy file
cp file.txt /path/to/destination/   # Copy to directory
cp -r dir1 dir2                     # Copy directory recursively
cp -i file.txt dest/                # Interactive (prompt before overwrite)
cp -v file.txt dest/                # Verbose output
cp -p file.txt dest/                # Preserve permissions and timestamps
cp -a dir1 dir2                     # Archive mode (preserve all attributes)
cp -u source dest                   # Copy only if source is newer
cp *.txt /backup/                   # Copy multiple files with wildcard
```

### `mv` - Move/Rename Files and Directories
Move or rename files and directories.

```bash
# Syntax
mv [options] source destination

# Examples
mv oldname.txt newname.txt          # Rename file
mv file.txt /new/path/              # Move file to new location
mv dir1 dir2                        # Rename or move directory
mv -i file.txt dest/                # Interactive (prompt before overwrite)
mv -v file.txt dest/                # Verbose output
mv -n file.txt dest/                # No overwrite if exists
mv -u file.txt dest/                # Move only if source is newer
mv *.log /archive/                  # Move multiple files
```

### `rm` - Remove Files and Directories
Delete files and directories.

```bash
# Syntax
rm [options] file/directory

# Examples
rm file.txt                     # Remove file
rm file1.txt file2.txt          # Remove multiple files
rm -i file.txt                  # Interactive (prompt before delete)
rm -f file.txt                  # Force delete (no prompt)
rm -r directory                 # Remove directory recursively
rm -rf directory                # Force remove directory (USE WITH CAUTION!)
rm -v file.txt                  # Verbose output
rm *.tmp                        # Remove files matching pattern

# WARNING: rm -rf / will delete everything! Always double-check paths!
```

### `find` - Search for Files
Powerful file searching command.

```bash
# Syntax
find [path] [options] [expression]

# Examples
find . -name "*.txt"                    # Find all .txt files in current dir
find / -name "config.yml"               # Find file by name from root
find . -iname "*.TXT"                   # Case-insensitive search
find . -type f                          # Find only files
find . -type d                          # Find only directories
find . -type l                          # Find only symbolic links
find . -size +100M                      # Files larger than 100MB
find . -size -1k                        # Files smaller than 1KB
find . -mtime -7                        # Modified in last 7 days
find . -mtime +30                       # Modified more than 30 days ago
find . -atime -1                        # Accessed in last 24 hours
find . -user devops                     # Files owned by user 'devops'
find . -group admin                     # Files owned by group 'admin'
find . -perm 755                        # Files with exact permissions
find . -perm -755                       # Files with at least these permissions
find . -empty                           # Find empty files/directories
find . -name "*.log" -delete            # Find and delete
find . -name "*.sh" -exec chmod +x {} \; # Find and execute command
find . -name "*.txt" -exec grep "error" {} \; # Search in found files
find . -maxdepth 2 -name "*.conf"       # Limit search depth
find . -name "*.log" | xargs rm         # Using xargs
find . -newer reference.txt             # Files newer than reference
find . -type f -name "*.tmp" -mtime +7 -delete  # Complex query
```

### `locate` - Quick File Search
Fast file search using database.

```bash
# Syntax
locate [options] pattern

# Examples
locate nginx.conf               # Find files containing 'nginx.conf'
locate -i README                # Case-insensitive search
locate -c "*.log"               # Count matching files
locate -l 10 "*.conf"           # Limit to 10 results
sudo updatedb                   # Update locate database
```

### `which` - Locate Command
Find the path of an executable.

```bash
# Syntax
which [options] command

# Examples
which python                    # Find python executable
which -a python                 # Show all matching executables
which docker                    # Output: /usr/bin/docker
```

### `whereis` - Locate Binary, Source, Manual
Find binary, source, and manual pages.

```bash
# Syntax
whereis [options] command

# Examples
whereis bash
# Output: bash: /bin/bash /usr/share/man/man1/bash.1.gz
whereis -b bash                 # Binary only
whereis -m bash                 # Manual only
```

## Links

### `ln` - Create Links
Create hard and symbolic links.

```bash
# Syntax
ln [options] target link_name

# Hard Link (same inode, can't cross filesystems)
ln file.txt hardlink.txt

# Symbolic Link (soft link, can cross filesystems)
ln -s /path/to/original /path/to/link
ln -s /var/log/nginx logs           # Create symlink 'logs' pointing to nginx logs

# Examples
ln -sf target link                  # Force create (overwrite existing)
ln -v source link                   # Verbose output
ls -l link                          # Check where link points to
readlink link                       # Show symlink target
readlink -f link                    # Show absolute path of symlink target
```

# 2. File Content & Text Processing

## Viewing File Contents

### `cat` - Concatenate and Display
Display file contents.

```bash
# Syntax
cat [options] filename

# Examples
cat file.txt                    # Display file content
cat file1.txt file2.txt         # Display multiple files
cat -n file.txt                 # Show line numbers
cat -b file.txt                 # Number non-blank lines only
cat -s file.txt                 # Squeeze blank lines
cat -A file.txt                 # Show all characters (tabs, line endings)
cat > newfile.txt               # Create file (Ctrl+D to save)
cat >> file.txt                 # Append to file
cat file1.txt file2.txt > combined.txt  # Concatenate files
```

### `less` - View File with Pagination
View files with forward/backward navigation.

```bash
# Syntax
less [options] filename

# Examples
less file.txt                   # Open file in less
less +G file.txt                # Start at end of file
less -N file.txt                # Show line numbers
less -S file.txt                # Don't wrap long lines

# Navigation in less:
# Space/f    - Next page
# b          - Previous page
# /pattern   - Search forward
# ?pattern   - Search backward
# n          - Next search result
# N          - Previous search result
# g          - Go to beginning
# G          - Go to end
# q          - Quit
```

### `more` - View File with Forward Pagination
Basic file viewer (forward only).

```bash
# Syntax
more [options] filename

# Examples
more file.txt                   # View file
more -10 file.txt               # Display 10 lines at a time
more +100 file.txt              # Start at line 100
```

### `head` - Display Beginning of File
Show first lines of a file.

```bash
# Syntax
head [options] filename

# Examples
head file.txt                   # First 10 lines (default)
head -n 20 file.txt             # First 20 lines
head -20 file.txt               # Same as above
head -c 100 file.txt            # First 100 bytes
head -n -5 file.txt             # All but last 5 lines
head file1.txt file2.txt        # Multiple files with headers
```

### `tail` - Display End of File
Show last lines of a file.

```bash
# Syntax
tail [options] filename

# Examples
tail file.txt                   # Last 10 lines (default)
tail -n 20 file.txt             # Last 20 lines
tail -20 file.txt               # Same as above
tail -c 100 file.txt            # Last 100 bytes
tail -n +5 file.txt             # Start from line 5
tail -f /var/log/syslog         # Follow file (real-time updates) - VERY USEFUL!
tail -F /var/log/app.log        # Follow and retry if file recreated
tail -f log1.log log2.log       # Follow multiple files
```

### `wc` - Word Count
Count lines, words, and characters.

```bash
# Syntax
wc [options] filename

# Examples
wc file.txt                     # Lines, words, characters
wc -l file.txt                  # Count lines only
wc -w file.txt                  # Count words only
wc -c file.txt                  # Count bytes
wc -m file.txt                  # Count characters
wc -L file.txt                  # Length of longest line
wc *.txt                        # Count for multiple files
cat file.txt | wc -l            # Count lines from stdin
```

## Text Searching

### `grep` - Search Text Patterns
Search for patterns in files - **Most Important Tool!**

```bash
# Syntax
grep [options] pattern [file...]

# Basic Examples
grep "error" file.txt               # Search for 'error' in file
grep "error" *.log                  # Search in multiple files
grep -i "error" file.txt            # Case-insensitive search
grep -v "debug" file.txt            # Invert match (exclude lines)
grep -n "error" file.txt            # Show line numbers
grep -c "error" file.txt            # Count matches
grep -l "error" *.log               # List files with matches
grep -L "error" *.log               # List files without matches

# Recursive Search
grep -r "TODO" .                    # Recursive search in directory
grep -R "config" /etc/              # Same as -r
grep -rn "password" .               # Recursive with line numbers

# Regular Expressions
grep -E "error|warning" file.txt    # Extended regex (egrep)
grep "^Start" file.txt              # Lines starting with 'Start'
grep "end$" file.txt                # Lines ending with 'end'
grep "^$" file.txt                  # Empty lines
grep "." file.txt                   # Non-empty lines
grep "[0-9]" file.txt               # Lines with numbers
grep "[A-Za-z]" file.txt            # Lines with letters
grep -E "[0-9]{3}-[0-9]{4}" file.txt # Pattern: 123-4567

# Context Options
grep -B 3 "error" file.txt          # 3 lines before match
grep -A 3 "error" file.txt          # 3 lines after match
grep -C 3 "error" file.txt          # 3 lines before and after

# Advanced Options
grep -w "error" file.txt            # Match whole word only
grep -x "exact line" file.txt       # Match exact line
grep -o "pattern" file.txt          # Print only matched part
grep --color=auto "error" file.txt  # Highlight matches
grep -P "\d{3}" file.txt            # Perl regex
grep -f patterns.txt file.txt       # Patterns from file
grep -e "pat1" -e "pat2" file.txt   # Multiple patterns

# Practical DevOps Examples
grep -r "Exception" /var/log/       # Find exceptions in logs
grep -E "(ERROR|WARN|FATAL)" app.log  # Multiple log levels
grep "Failed password" /var/log/auth.log  # Security audit
dmesg | grep -i "error"             # Kernel errors
docker logs container | grep "error"  # Docker log filtering
kubectl logs pod-name | grep -i "exception"  # K8s logs
```

### `egrep` / `fgrep` - Extended/Fixed Grep
```bash
# egrep - Extended regex (same as grep -E)
egrep "error|warning|critical" file.txt

# fgrep - Fixed strings, no regex (same as grep -F)
fgrep "exact$tring" file.txt        # Treats $ literally
```

## Text Manipulation

### `cut` - Extract Columns
Extract sections from lines.

```bash
# Syntax
cut [options] filename

# Examples
cut -d ":" -f 1 /etc/passwd         # Extract first field (delimiter :)
cut -d "," -f 1,3 data.csv          # Extract fields 1 and 3
cut -d " " -f 2- file.txt           # Fields 2 to end
cut -c 1-10 file.txt                # Characters 1-10
cut -c 5 file.txt                   # Only 5th character
cut -c -5 file.txt                  # First 5 characters
cut -c 5- file.txt                  # From 5th character to end
echo "hello:world" | cut -d ":" -f 2  # Output: world
```

### `sort` - Sort Lines
Sort file contents.

```bash
# Syntax
sort [options] filename

# Examples
sort file.txt                       # Alphabetical sort
sort -r file.txt                    # Reverse sort
sort -n numbers.txt                 # Numeric sort
sort -nr numbers.txt                # Reverse numeric sort
sort -u file.txt                    # Unique (remove duplicates)
sort -k 2 file.txt                  # Sort by 2nd column
sort -k 2,2 file.txt                # Sort by 2nd column only
sort -t ":" -k 3 -n /etc/passwd     # Sort passwd by UID
sort -k 2 -t "," -n data.csv        # Sort CSV by 2nd column numerically
sort -h sizes.txt                   # Human numeric sort (1K, 2M, 3G)
sort -M dates.txt                   # Month sort (Jan, Feb...)
sort -f file.txt                    # Ignore case
sort file1.txt file2.txt > sorted.txt  # Merge and sort
sort -o output.txt input.txt        # Sort in place
```

### `uniq` - Remove Duplicates
Filter adjacent matching lines.

```bash
# Syntax
uniq [options] [input [output]]

# Examples (NOTE: file must be sorted first!)
sort file.txt | uniq                # Remove duplicates
sort file.txt | uniq -c             # Count occurrences
sort file.txt | uniq -d             # Show only duplicates
sort file.txt | uniq -u             # Show only unique lines
sort file.txt | uniq -i             # Ignore case
sort access.log | uniq -c | sort -rn | head -10  # Top 10 most frequent
```

### `tr` - Translate Characters
Replace or delete characters.

```bash
# Syntax
tr [options] SET1 [SET2]

# Examples
echo "hello" | tr 'a-z' 'A-Z'       # Convert to uppercase
echo "HELLO" | tr 'A-Z' 'a-z'       # Convert to lowercase
echo "hello world" | tr ' ' '_'     # Replace spaces with underscores
echo "hello123" | tr -d '0-9'       # Delete digits: hello
echo "aabbcc" | tr -s 'a-z'         # Squeeze repeated: abc
cat file.txt | tr -d '\r'           # Remove Windows line endings
echo "hello" | tr -c 'aeiou' '-'    # Complement: -e--o (non-vowels to -)
tr ':' '\n' <<< "$PATH"             # Display PATH directories
```

### `diff` - Compare Files
Show differences between files.

```bash
# Syntax
diff [options] file1 file2

# Examples
diff file1.txt file2.txt            # Show differences
diff -u file1.txt file2.txt         # Unified format (like git diff)
diff -c file1.txt file2.txt         # Context format
diff -y file1.txt file2.txt         # Side-by-side comparison
diff -q file1.txt file2.txt         # Quick check (only if different)
diff -r dir1/ dir2/                 # Compare directories recursively
diff -i file1.txt file2.txt         # Ignore case differences
diff -w file1.txt file2.txt         # Ignore whitespace
diff -B file1.txt file2.txt         # Ignore blank lines
diff --color file1.txt file2.txt    # Colorized output

# Create patch file
diff -u original.txt modified.txt > changes.patch
# Apply patch
patch original.txt < changes.patch
```

### `sed` - Stream Editor (Basic)
Process and transform text.

```bash
# Syntax
sed [options] 'command' filename

# Basic Substitution
sed 's/old/new/' file.txt           # Replace first occurrence per line
sed 's/old/new/g' file.txt          # Replace all occurrences (global)
sed 's/old/new/gi' file.txt         # Global, case-insensitive
sed 's/old/new/2' file.txt          # Replace 2nd occurrence only
sed 's|/usr/bin|/usr/local/bin|g' file.txt  # Use different delimiter

# Line Operations
sed '3d' file.txt                   # Delete line 3
sed '2,5d' file.txt                 # Delete lines 2-5
sed '/pattern/d' file.txt           # Delete lines matching pattern
sed '/^$/d' file.txt                # Delete empty lines
sed '/^#/d' file.txt                # Delete comment lines
sed '3p' file.txt                   # Print line 3 (with -n)
sed -n '5,10p' file.txt             # Print lines 5-10 only
sed '2a\New line' file.txt          # Append after line 2
sed '2i\New line' file.txt          # Insert before line 2
sed '2c\Replaced line' file.txt     # Replace line 2

# In-place Editing
sed -i 's/old/new/g' file.txt       # Edit file in place
sed -i.bak 's/old/new/g' file.txt   # In place with backup
sed -i '' 's/old/new/g' file.txt    # In place without backup (macOS)

# Practical Examples
sed 's/^/    /' file.txt            # Add 4 spaces to beginning
sed 's/[[:space:]]*$//' file.txt   # Remove trailing whitespace
sed '1d' file.txt                   # Remove first line (header)
sed '$d' file.txt                   # Remove last line
sed 's/<[^>]*>//g' file.html        # Remove HTML tags
```

# 3. User & Permission Management

## User Management

### `whoami` - Current User
Display current username.

```bash
whoami
# Output: devops
```

### `id` - User Identity
Display user and group IDs.

```bash
# Syntax
id [options] [username]

# Examples
id                      # Current user's UID, GID, groups
id devops               # Specific user's info
id -u                   # User ID only
id -g                   # Primary group ID only
id -G                   # All group IDs
id -un                  # Username
id -gn                  # Primary group name
id -Gn                  # All group names
```

### `useradd` - Create User
Create a new user account.

```bash
# Syntax
useradd [options] username

# Examples
sudo useradd john                           # Create user (minimal)
sudo useradd -m john                        # Create with home directory
sudo useradd -m -s /bin/bash john           # With home dir and shell
sudo useradd -m -d /home/john -s /bin/bash john  # Custom home dir
sudo useradd -G sudo,docker john            # Add to secondary groups
sudo useradd -u 1500 john                   # Specify UID
sudo useradd -g developers john             # Specify primary group
sudo useradd -c "John Doe" john             # Add comment/full name
sudo useradd -e 2025-12-31 john             # Account expiry date
sudo useradd -M john                        # No home directory

# Common DevOps User Creation
sudo useradd -m -s /bin/bash -G sudo,docker,wheel devops
```

### `usermod` - Modify User
Modify existing user account.

```bash
# Syntax
usermod [options] username

# Examples
sudo usermod -aG sudo john              # Add to group (keep existing groups)
sudo usermod -aG docker john            # Add to docker group
sudo usermod -g developers john         # Change primary group
sudo usermod -s /bin/zsh john           # Change shell
sudo usermod -d /new/home john          # Change home directory
sudo usermod -l newname oldname         # Rename user
sudo usermod -L john                    # Lock account
sudo usermod -U john                    # Unlock account
sudo usermod -e 2025-12-31 john         # Set expiry date
sudo usermod -e "" john                 # Remove expiry date
```

### `userdel` - Delete User
Remove a user account.

```bash
# Syntax
userdel [options] username

# Examples
sudo userdel john                       # Delete user (keep home dir)
sudo userdel -r john                    # Delete user and home directory
sudo userdel -f john                    # Force delete (even if logged in)
```

### `passwd` - Change Password
Set or change user password.

```bash
# Syntax
passwd [options] [username]

# Examples
passwd                                  # Change own password
sudo passwd john                        # Change john's password
sudo passwd -l john                     # Lock account
sudo passwd -u john                     # Unlock account
sudo passwd -d john                     # Delete password
sudo passwd -e john                     # Expire password (force change)
sudo passwd -S john                     # Show password status
chage -l john                           # Show password aging info
```

### `groupadd` / `groupdel` / `groupmod` - Group Management
Manage user groups.

```bash
# Create Group
sudo groupadd developers
sudo groupadd -g 1500 developers        # With specific GID

# Delete Group
sudo groupdel developers

# Modify Group
sudo groupmod -n newname oldname        # Rename group
sudo groupmod -g 1501 developers        # Change GID

# List Groups
groups                                  # Current user's groups
groups john                             # Specific user's groups
cat /etc/group                          # All groups
getent group developers                 # Group info
```

### `su` - Switch User
Switch to another user.

```bash
# Syntax
su [options] [username]

# Examples
su john                         # Switch to john (keep environment)
su - john                       # Switch with login shell (new environment)
su -                            # Switch to root with login shell
su -c "command" john            # Run single command as john
sudo su -                       # Switch to root using sudo
```

### `sudo` - Superuser Do
Execute commands as superuser.

```bash
# Syntax
sudo [options] command

# Examples
sudo apt update                         # Run as root
sudo -u john command                    # Run as specific user
sudo -i                                 # Interactive root shell
sudo -s                                 # Shell as root (keep env)
sudo su -                               # Full root login shell
sudo -l                                 # List allowed commands
sudo -v                                 # Update cached credentials
sudo -k                                 # Invalidate cached credentials
sudo visudo                             # Edit sudoers file safely

# /etc/sudoers examples:
# john ALL=(ALL:ALL) ALL                # Full sudo access
# john ALL=(ALL) NOPASSWD: ALL          # Sudo without password
# %sudo ALL=(ALL:ALL) ALL               # Group sudo access
```

## File Permissions

### `chmod` - Change Permissions
Modify file permissions.

```bash
# Syntax
chmod [options] mode file

# Permission Notation:
# r (read)    = 4
# w (write)   = 2
# x (execute) = 1
# 
# Owner | Group | Others
# rwx   | rwx   | rwx
# 7     | 5     | 5     = 755

# Numeric Mode Examples
chmod 755 script.sh         # rwxr-xr-x (common for executables)
chmod 644 file.txt          # rw-r--r-- (common for files)
chmod 600 private.key       # rw------- (private files)
chmod 700 directory         # rwx------ (private directory)
chmod 777 file.txt          # rwxrwxrwx (full access - avoid!)
chmod 666 file.txt          # rw-rw-rw- (read-write all)
chmod 400 readonly.txt      # r-------- (read-only owner)

# Symbolic Mode Examples
chmod +x script.sh          # Add execute for all
chmod -x script.sh          # Remove execute for all
chmod u+x script.sh         # Add execute for owner (user)
chmod g+w file.txt          # Add write for group
chmod o-r file.txt          # Remove read for others
chmod u+rwx,g+rx,o+r file   # Multiple permissions
chmod a+r file.txt          # Add read for all (a = all)
chmod u=rwx,g=rx,o=r file   # Set exact permissions

# Recursive Permission Change
chmod -R 755 directory/     # Apply to all contents recursively
chmod -R +x scripts/        # Make all files executable

# Special Permissions
chmod 4755 file             # SetUID (run as owner)
chmod 2755 directory        # SetGID (inherit group)
chmod 1777 /tmp             # Sticky bit (only owner can delete)
chmod u+s file              # SetUID symbolic
chmod g+s directory         # SetGID symbolic
chmod +t directory          # Sticky bit symbolic

# Options
chmod -v 755 file           # Verbose
chmod -c 755 file           # Report only changes
chmod --reference=ref file  # Copy permissions from reference
```

### `chown` - Change Ownership
Change file owner and group.

```bash
# Syntax
chown [options] owner[:group] file

# Examples
sudo chown john file.txt            # Change owner
sudo chown john:developers file.txt # Change owner and group
sudo chown :developers file.txt     # Change group only
sudo chown john: file.txt           # Change owner, group to owner's login group
sudo chown -R john:developers dir/  # Recursive change
sudo chown --reference=ref file     # Copy ownership from reference

# Practical Examples
sudo chown www-data:www-data /var/www/html  # Web server files
sudo chown -R $USER:$USER ~/project         # Take ownership
sudo chown root:root /etc/shadow            # System file ownership
```

### `chgrp` - Change Group
Change file group ownership.

```bash
# Syntax
chgrp [options] group file

# Examples
sudo chgrp developers file.txt      # Change group
sudo chgrp -R developers dir/       # Recursive change
sudo chgrp --reference=ref file     # Copy group from reference
```

### `umask` - Default Permissions
Set default permission mask for new files.

```bash
# Syntax
umask [mode]

# Understanding umask:
# Files default: 666 (rw-rw-rw-)
# Dirs default:  777 (rwxrwxrwx)
# umask subtracts from these defaults

# Examples
umask                   # Show current umask (e.g., 0022)
umask 022               # Files: 644, Dirs: 755 (common default)
umask 077               # Files: 600, Dirs: 700 (restrictive)
umask 002               # Files: 664, Dirs: 775 (group-friendly)

# Calculation:
# umask 022: Files = 666 - 022 = 644, Dirs = 777 - 022 = 755
# umask 077: Files = 666 - 077 = 600, Dirs = 777 - 077 = 700
```

### `getfacl` / `setfacl` - Access Control Lists
Advanced permission management.

```bash
# View ACL
getfacl file.txt

# Set ACL - Give user specific permissions
setfacl -m u:john:rwx file.txt          # User john gets rwx
setfacl -m g:developers:rx file.txt     # Group developers gets rx
setfacl -m o::r file.txt                # Others get r

# Default ACL for directories (inherited by new files)
setfacl -d -m u:john:rwx directory/
setfacl -d -m g:developers:rx directory/

# Remove ACL
setfacl -x u:john file.txt              # Remove user ACL
setfacl -b file.txt                     # Remove all ACL

# Recursive ACL
setfacl -R -m u:john:rx directory/
```

```bash
# Syntax
kill [options] PID

# Common Signals:
# SIGTERM (15) - Graceful termination (default)
# SIGKILL (9)  - Force kill (cannot be caught)
# SIGHUP (1)   - Hangup (often used to reload config)
# SIGSTOP (19) - Pause process
# SIGCONT (18) - Continue paused process
# SIGINT (2)   - Interrupt (like Ctrl+C)

# Examples
kill 1234                   # Send SIGTERM to PID 1234
kill -15 1234               # Same as above
kill -TERM 1234             # Same as above
kill -9 1234                # Force kill (SIGKILL)
kill -KILL 1234             # Same as above
kill -1 1234                # Send SIGHUP (reload)
kill -HUP 1234              # Same as above

# Kill Multiple Processes
kill 1234 5678 9012         # Kill multiple PIDs
kill $(pgrep nginx)         # Kill all nginx processes

# List Available Signals
kill -l                     # List all signals
```

```bash
# Syntax
tar [options] archive_name files

# Common options:
# -c  Create archive
# -x  Extract archive
# -t  List contents
# -v  Verbose
# -f  File name
# -z  gzip compression
# -j  bzip2 compression
# -J  xz compression

# Create Archives
tar -cvf archive.tar files/         # Create tar archive
tar -cvf archive.tar file1 file2    # Multiple files
tar -cvzf archive.tar.gz files/     # Create gzip compressed
tar -cvjf archive.tar.bz2 files/    # Create bzip2 compressed
tar -cvJf archive.tar.xz files/     # Create xz compressed

# Extract Archives
tar -xvf archive.tar                # Extract tar
tar -xvf archive.tar -C /dest/      # Extract to directory
tar -xvzf archive.tar.gz            # Extract gzip
tar -xvjf archive.tar.bz2           # Extract bzip2
tar -xvJf archive.tar.xz            # Extract xz

# List Contents
tar -tvf archive.tar                # List contents
tar -tvzf archive.tar.gz            # List gzip archive

# Append/Update
tar -rvf archive.tar newfile        # Append file
tar -uvf archive.tar files/         # Update changed files

# Exclude files
tar -cvzf archive.tar.gz --exclude="*.log" dir/
tar -cvzf archive.tar.gz --exclude-vcs dir/  # Exclude .git, .svn

# Practical examples
tar -cvzf backup-$(date +%Y%m%d).tar.gz /home/user
tar -xvzf archive.tar.gz --strip-components=1  # Remove top directory
```

## Compression Tools

### `gzip` / `gunzip` - GNU Zip
```bash
# Compress
gzip file.txt                       # Creates file.txt.gz (removes original)
gzip -k file.txt                    # Keep original file
gzip -1 file.txt                    # Fast compression
gzip -9 file.txt                    # Best compression
gzip -r directory/                  # Recursive

# Decompress
gunzip file.txt.gz                  # Decompress
gzip -d file.txt.gz                 # Same as gunzip
gzip -l file.txt.gz                 # List compression info
zcat file.txt.gz                    # View without decompressing
```

### `bzip2` / `bunzip2` - Better Compression
```bash
# Compress
bzip2 file.txt                      # Creates file.txt.bz2
bzip2 -k file.txt                   # Keep original
bzip2 -9 file.txt                   # Best compression

# Decompress
bunzip2 file.txt.bz2                # Decompress
bzip2 -d file.txt.bz2               # Same as bunzip2
bzcat file.txt.bz2                  # View without decompressing
```

### `xz` - High Compression
```bash
# Compress
xz file.txt                         # Creates file.txt.xz
xz -k file.txt                      # Keep original
xz -9 file.txt                      # Best compression
xz -T 4 file.txt                    # Use 4 threads

# Decompress
unxz file.txt.xz                    # Decompress
xz -d file.txt.xz                   # Same as unxz
xzcat file.txt.xz                   # View without decompressing
```

### `zip` / `unzip` - ZIP Format
```bash
# Create ZIP
zip archive.zip file1 file2         # Create zip
zip -r archive.zip directory/       # Recursive
zip -9 archive.zip files            # Best compression
zip -e archive.zip files            # Encrypt with password
zip -u archive.zip newfile          # Update/add file

# Extract ZIP
unzip archive.zip                   # Extract
unzip archive.zip -d /dest/         # Extract to directory
unzip -l archive.zip                # List contents
unzip -o archive.zip                # Overwrite without prompt
unzip -n archive.zip                # Never overwrite
```

## SSH - Secure Shell

### Basic SSH Connection
```bash
# Syntax
ssh [options] user@host

# Examples
ssh user@192.168.1.100              # Connect to server
ssh -p 2222 user@host               # Custom port
ssh user@host command               # Run remote command
ssh -i ~/.ssh/mykey.pem user@host   # Use specific key
ssh -v user@host                    # Verbose (debug)
ssh -vv user@host                   # More verbose
ssh -o StrictHostKeyChecking=no user@host  # Skip key check

# Run command and exit
ssh user@host "ls -la /var/log"
ssh user@host "df -h && free -m"

# Interactive with pseudo-terminal
ssh -t user@host "top"              # Force TTY allocation
ssh -t user@host "sudo systemctl restart nginx"
```

### SSH Key Management
```bash
# Generate SSH key pair
ssh-keygen                                      # Default RSA
ssh-keygen -t rsa -b 4096                       # RSA 4096-bit
ssh-keygen -t ed25519                           # ED25519 (recommended)
ssh-keygen -t ed25519 -C "email@example.com"    # With comment
ssh-keygen -f ~/.ssh/mykey                      # Custom filename

# Key files:
# ~/.ssh/id_rsa         - Private key
# ~/.ssh/id_rsa.pub     - Public key
# ~/.ssh/known_hosts    - Known hosts
# ~/.ssh/authorized_keys - Authorized keys (server)

# Copy public key to server
ssh-copy-id user@host                           # Copy default key
ssh-copy-id -i ~/.ssh/mykey.pub user@host       # Specific key
ssh-copy-id -p 2222 user@host                   # Custom port

# Manual key copy (if ssh-copy-id unavailable)
cat ~/.ssh/id_rsa.pub | ssh user@host "mkdir -p ~/.ssh && cat >> ~/.ssh/authorized_keys"

# SSH Agent (store keys in memory)
eval $(ssh-agent)                   # Start agent
ssh-add                             # Add default key
ssh-add ~/.ssh/mykey                # Add specific key
ssh-add -l                          # List loaded keys
ssh-add -d ~/.ssh/mykey             # Remove key
```

### SSH Config File
```bash
# ~/.ssh/config
# Makes SSH connections easier with aliases

Host myserver
    HostName 192.168.1.100
    User devops
    Port 22
    IdentityFile ~/.ssh/mykey

Host prod-*
    User admin
    IdentityFile ~/.ssh/prod_key

Host *
    ServerAliveInterval 60
    ServerAliveCountMax 3

# Usage:
ssh myserver                        # Instead of: ssh devops@192.168.1.100
```

### SSH Tunneling / Port Forwarding
```bash
# Local Port Forwarding (access remote service locally)
# Forward local port 8080 to remote server's port 80
ssh -L 8080:localhost:80 user@remotehost
# Now access localhost:8080 to reach remotehost:80

# Access database through SSH tunnel
ssh -L 3306:db.internal:3306 user@jumphost
# Connect to localhost:3306 to access db.internal:3306

# Remote Port Forwarding (expose local service remotely)
ssh -R 8080:localhost:80 user@remotehost
# Remote server's port 8080 forwards to your localhost:80

# Dynamic Port Forwarding (SOCKS proxy)
ssh -D 1080 user@host
# Configure browser to use SOCKS proxy localhost:1080

# Keep tunnel alive in background
ssh -fN -L 8080:localhost:80 user@host
# -f: Background
# -N: No command (just tunnel)

# Jump Host / Bastion
ssh -J jumphost user@finalhost
ssh -o ProxyJump=jumphost user@finalhost
```

# 18. AWK & SED - Text Processing

## AWK - Pattern Scanning

### Basic AWK Usage
```bash
# Syntax
awk 'pattern {action}' file

# Print entire file
awk '{print}' file.txt
awk '{print $0}' file.txt               # Same as above

# Print specific columns
awk '{print $1}' file.txt               # First column
awk '{print $1, $3}' file.txt           # First and third columns
awk '{print $NF}' file.txt              # Last column
awk '{print $(NF-1)}' file.txt          # Second to last

# Field separator
awk -F ':' '{print $1}' /etc/passwd     # Colon separator
awk -F ',' '{print $1, $2}' data.csv    # CSV file
awk 'BEGIN{FS=":"} {print $1}' /etc/passwd

# Pattern matching
awk '/error/' file.txt                  # Lines containing 'error'
awk '/^start/' file.txt                 # Lines starting with 'start'
awk '/end$/' file.txt                   # Lines ending with 'end'
awk '!/comment/' file.txt               # Lines NOT containing 'comment'

# Conditions
awk '$3 > 100' file.txt                 # Third column > 100
awk '$1 == "admin"' file.txt            # First column equals 'admin'
awk 'NR > 1' file.txt                   # Skip header (line 2+)
awk 'NR >= 5 && NR <= 10' file.txt      # Lines 5-10
awk 'length($0) > 80' file.txt          # Lines longer than 80 chars

# Built-in variables
awk '{print NR, $0}' file.txt           # NR = line number
awk '{print NF}' file.txt               # NF = number of fields
awk 'END{print NR}' file.txt            # Total lines
awk '{print FILENAME, $0}' file.txt     # FILENAME = current file

# Calculations
awk '{sum += $1} END{print sum}' file.txt           # Sum column
awk '{sum += $1} END{print sum/NR}' file.txt        # Average
awk 'BEGIN{max=0} $1>max{max=$1} END{print max}' file.txt  # Maximum
awk '{print $1, $2, $1+$2}' file.txt                # Add columns

# Formatting
awk '{printf "%-10s %5d\n", $1, $2}' file.txt       # Formatted output
awk '{printf "%s: $%.2f\n", $1, $2}' file.txt       # Currency format

# BEGIN and END blocks
awk 'BEGIN{print "Header"} {print} END{print "Footer"}' file.txt
awk 'BEGIN{count=0} /error/{count++} END{print count}' file.txt

# Practical examples
# Parse access logs
awk '{print $1}' access.log | sort | uniq -c | sort -rn | head -10

# Sum file sizes
ls -l | awk '{sum += $5} END{print sum}'

# Parse /etc/passwd
awk -F ':' '$3 >= 1000 {print $1}' /etc/passwd      # Regular users

# CSV processing
awk -F ',' 'NR>1 {sum+=$3} END{print "Total:", sum}' data.csv
```

## SED - Stream Editor (Advanced)

### Advanced SED Usage
```bash
# Syntax
sed [options] 'command' file

# Substitution
sed 's/old/new/' file                   # First occurrence per line
sed 's/old/new/g' file                  # All occurrences (global)
sed 's/old/new/2' file                  # Second occurrence only
sed 's/old/new/gi' file                 # Global, case-insensitive
sed 's|/path/old|/path/new|g' file      # Different delimiter

# Multiple substitutions
sed -e 's/foo/bar/g' -e 's/baz/qux/g' file
sed 's/foo/bar/g; s/baz/qux/g' file

# Addressing
sed '3s/old/new/' file                  # Only line 3
sed '2,5s/old/new/' file                # Lines 2-5
sed '5,$s/old/new/' file                # Line 5 to end
sed '/pattern/s/old/new/' file          # Lines matching pattern
sed '/start/,/end/s/old/new/' file      # Range between patterns

# Line operations
sed '5d' file                           # Delete line 5
sed '2,5d' file                         # Delete lines 2-5
sed '/pattern/d' file                   # Delete matching lines
sed '/^$/d' file                        # Delete empty lines
sed '/^#/d' file                        # Delete comments
sed '1d' file                           # Delete first line
sed '$d' file                           # Delete last line

# Insert and append
sed '3i\New line before' file           # Insert before line 3
sed '3a\New line after' file            # Append after line 3
sed '/pattern/i\New line' file          # Insert before pattern
sed '/pattern/a\New line' file          # Append after pattern

# Print
sed -n '5p' file                        # Print line 5 only
sed -n '2,5p' file                      # Print lines 2-5
sed -n '/pattern/p' file                # Print matching lines
sed -n '1p;$p' file                     # First and last lines

# In-place editing
sed -i 's/old/new/g' file               # Edit file directly
sed -i.bak 's/old/new/g' file           # With backup

# Practical examples
# Remove Windows line endings
sed 's/\r$//' file > file.unix

# Remove leading/trailing whitespace
sed 's/^[ \t]*//' file                  # Leading
sed 's/[ \t]*$//' file                  # Trailing
sed 's/^[ \t]*//;s/[ \t]*$//' file      # Both

# Add line numbers
sed = file | sed 'N;s/\n/\t/'

# Replace newlines
sed ':a;N;$!ba;s/\n/ /g' file

# Extract between patterns
sed -n '/START/,/END/p' file

# Comment/uncomment lines
sed 's/^/#/' file                       # Comment all
sed 's/^#//' file                       # Uncomment all
sed '/pattern/s/^/#/' file              # Comment matching lines
```

## Combining AWK & SED

```bash
# Pipeline examples
# Parse and format logs
cat access.log | awk '{print $1, $7}' | sed 's/HTTP.*//' | sort | uniq -c

# Extract and transform data
cat data.csv | sed '1d' | awk -F ',' '{print $1, $3}' | sort -t ' ' -k2 -rn

# Clean and analyze
grep "ERROR" app.log | sed 's/.*ERROR//' | awk '{print $1}' | sort | uniq -c

# Multi-step transformation
cat /etc/passwd | \
    awk -F ':' '{print $1, $3}' | \
    sed 's/ /: UID /' | \
    sort -t ' ' -k3 -n
```

# 19. Cron Jobs & Scheduling

## Cron - Task Scheduling

### Cron Syntax
```bash
# Crontab format:
# ┌───────────── minute (0 - 59)
# │ ┌───────────── hour (0 - 23)
# │ │ ┌───────────── day of month (1 - 31)
# │ │ │ ┌───────────── month (1 - 12)
# │ │ │ │ ┌───────────── day of week (0 - 7) (Sunday = 0 or 7)
# │ │ │ │ │
# * * * * * command

# Special characters:
# *     Any value
# ,     Value list (1,3,5)
# -     Range (1-5)
# /     Step values (*/15 = every 15)
```

### Crontab Commands
```bash
# Edit crontab
crontab -e                              # Edit current user's crontab
sudo crontab -e                         # Edit root's crontab
crontab -u user -e                      # Edit another user's crontab

# List crontab
crontab -l                              # List current user's crontab
crontab -u user -l                      # List another user's crontab

# Remove crontab
crontab -r                              # Remove current user's crontab
crontab -ri                             # Remove with confirmation
```

### Cron Examples
```bash
# Every minute
* * * * * /path/to/script.sh

# Every 5 minutes
*/5 * * * * /path/to/script.sh

# Every hour at minute 30
30 * * * * /path/to/script.sh

# Every day at 2:30 AM
30 2 * * * /path/to/script.sh

# Every Monday at 3:00 AM
0 3 * * 1 /path/to/script.sh

# Every weekday at 9:00 AM
0 9 * * 1-5 /path/to/script.sh

# First day of every month at midnight
0 0 1 * * /path/to/script.sh

# Every 15 minutes during business hours
*/15 9-17 * * 1-5 /path/to/script.sh

# Twice a day at 6 AM and 6 PM
0 6,18 * * * /path/to/script.sh

# Every Sunday at 11:30 PM
30 23 * * 0 /path/to/script.sh

# Quarterly (Jan, Apr, Jul, Oct) on 1st at midnight
0 0 1 1,4,7,10 * /path/to/script.sh
```

### Special Cron Strings
```bash
@reboot         # Run once at startup
@yearly         # 0 0 1 1 * (once a year)
@annually       # Same as @yearly
@monthly        # 0 0 1 * * (once a month)
@weekly         # 0 0 * * 0 (once a week)
@daily          # 0 0 * * * (once a day)
@midnight       # Same as @daily
@hourly         # 0 * * * * (once an hour)

# Examples
@reboot /path/to/startup-script.sh
@daily /path/to/backup.sh
@hourly /path/to/health-check.sh
```

### Cron Best Practices
```bash
# Redirect output to log
* * * * * /path/to/script.sh >> /var/log/myjob.log 2>&1

# Redirect output to null (silent)
* * * * * /path/to/script.sh > /dev/null 2>&1

# Use full paths
0 2 * * * /usr/bin/python3 /home/user/script.py

# Set environment variables
SHELL=/bin/bash
PATH=/usr/local/sbin:/usr/local/bin:/sbin:/bin:/usr/sbin:/usr/bin
MAILTO=admin@example.com

# Lock to prevent overlap
* * * * * flock -n /tmp/myjob.lock /path/to/script.sh

# With timeout
* * * * * timeout 300 /path/to/script.sh
```

### System Cron Directories
```bash
/etc/crontab            # System crontab
/etc/cron.d/            # Additional cron files
/etc/cron.hourly/       # Hourly scripts
/etc/cron.daily/        # Daily scripts
/etc/cron.weekly/       # Weekly scripts
/etc/cron.monthly/      # Monthly scripts

# Example /etc/crontab format (includes user field):
# m h dom mon dow user command
0 2 * * * root /usr/local/bin/backup.sh
```

## at - One-time Scheduling

```bash
# Schedule one-time job
at 2:00 AM                              # At specific time
at 2:00 AM tomorrow                     # Tomorrow
at now + 2 hours                        # Relative time
at noon                                 # At noon
at midnight                             # At midnight

# Interactive mode
at 2:00 AM
> /path/to/script.sh
> Ctrl+D

# From command line
echo "/path/to/script.sh" | at 2:00 AM
at 2:00 AM -f /path/to/script.sh

# List scheduled jobs
atq
at -l

# Remove job
atrm job_number
at -d job_number

# View job details
at -c job_number
```

# 20. Security & Firewall

## iptables - Packet Filtering

### Basic iptables
```bash
# List rules
sudo iptables -L                        # List all rules
sudo iptables -L -v                     # Verbose
sudo iptables -L -n                     # Numeric (no DNS)
sudo iptables -L --line-numbers         # With line numbers
sudo iptables -S                        # Rules as commands

# Default policies
sudo iptables -P INPUT ACCEPT           # Accept all incoming
sudo iptables -P INPUT DROP             # Drop all incoming
sudo iptables -P OUTPUT ACCEPT          # Accept all outgoing
sudo iptables -P FORWARD DROP           # Drop forwarding

# Allow established connections
sudo iptables -A INPUT -m state --state ESTABLISHED,RELATED -j ACCEPT

# Allow specific ports
sudo iptables -A INPUT -p tcp --dport 22 -j ACCEPT      # SSH
sudo iptables -A INPUT -p tcp --dport 80 -j ACCEPT      # HTTP
sudo iptables -A INPUT -p tcp --dport 443 -j ACCEPT     # HTTPS

# Allow from specific IP
sudo iptables -A INPUT -s 192.168.1.100 -j ACCEPT
sudo iptables -A INPUT -s 192.168.1.0/24 -j ACCEPT      # Subnet

# Block specific IP
sudo iptables -A INPUT -s 10.0.0.5 -j DROP

# Allow ping
sudo iptables -A INPUT -p icmp --icmp-type echo-request -j ACCEPT

# Allow loopback
sudo iptables -A INPUT -i lo -j ACCEPT
sudo iptables -A OUTPUT -o lo -j ACCEPT

# Delete rules
sudo iptables -D INPUT 3                # Delete rule by number
sudo iptables -D INPUT -p tcp --dport 80 -j ACCEPT      # Delete specific

# Flush rules
sudo iptables -F                        # Flush all rules
sudo iptables -F INPUT                  # Flush INPUT chain

# Save/restore rules
sudo iptables-save > /etc/iptables.rules
sudo iptables-restore < /etc/iptables.rules

# NAT (Network Address Translation)
sudo iptables -t nat -A POSTROUTING -o eth0 -j MASQUERADE
sudo iptables -t nat -A PREROUTING -p tcp --dport 80 -j REDIRECT --to-port 8080
```

## UFW - Uncomplicated Firewall (Ubuntu)

### Basic UFW Commands
```bash
# Enable/Disable
sudo ufw enable                         # Enable firewall
sudo ufw disable                        # Disable firewall
sudo ufw status                         # Status
sudo ufw status verbose                 # Detailed status
sudo ufw status numbered                # With rule numbers

# Default policies
sudo ufw default deny incoming          # Deny incoming
sudo ufw default allow outgoing         # Allow outgoing

# Allow rules
sudo ufw allow 22                       # Allow port 22
sudo ufw allow ssh                      # Allow SSH by name
sudo ufw allow 80/tcp                   # Allow TCP port 80
sudo ufw allow 80,443/tcp               # Multiple ports
sudo ufw allow 6000:6007/tcp            # Port range
sudo ufw allow from 192.168.1.100       # Allow from IP
sudo ufw allow from 192.168.1.0/24      # Allow from subnet
sudo ufw allow from 192.168.1.100 to any port 22  # IP to specific port
sudo ufw allow in on eth0 to any port 80  # On specific interface

# Deny rules
sudo ufw deny 23                        # Deny port 23
sudo ufw deny from 10.0.0.5             # Deny from IP
sudo ufw deny in on eth0                # Deny on interface

# Delete rules
sudo ufw delete allow 80                # Delete by rule
sudo ufw delete 3                       # Delete by number
sudo ufw --force reset                  # Reset all rules

# Logging
sudo ufw logging on                     # Enable logging
sudo ufw logging medium                 # Set log level
# Logs at: /var/log/ufw.log

# Application profiles
sudo ufw app list                       # List available profiles
sudo ufw allow 'Nginx Full'             # Allow by profile
sudo ufw allow 'OpenSSH'
```

## firewalld - Dynamic Firewall (RHEL/CentOS)

### firewalld Commands
```bash
# Status
sudo systemctl start firewalld
sudo systemctl enable firewalld
sudo firewall-cmd --state               # Running state
sudo firewall-cmd --reload              # Reload rules

# Zones
sudo firewall-cmd --get-default-zone    # Default zone
sudo firewall-cmd --get-active-zones    # Active zones
sudo firewall-cmd --list-all            # List all rules
sudo firewall-cmd --list-all --zone=public

# Allow services
sudo firewall-cmd --add-service=http                    # Temporary
sudo firewall-cmd --add-service=http --permanent        # Permanent
sudo firewall-cmd --add-service=https --permanent
sudo firewall-cmd --remove-service=http --permanent
sudo firewall-cmd --list-services                       # List services

# Allow ports
sudo firewall-cmd --add-port=8080/tcp --permanent
sudo firewall-cmd --add-port=8080-8090/tcp --permanent  # Range
sudo firewall-cmd --remove-port=8080/tcp --permanent
sudo firewall-cmd --list-ports

# Rich rules
sudo firewall-cmd --add-rich-rule='rule family="ipv4" source address="192.168.1.100" accept' --permanent
sudo firewall-cmd --add-rich-rule='rule family="ipv4" source address="10.0.0.0/24" port port="22" protocol="tcp" accept' --permanent

# Port forwarding
sudo firewall-cmd --add-forward-port=port=80:proto=tcp:toport=8080 --permanent
sudo firewall-cmd --add-masquerade --permanent

# Always reload after permanent changes
sudo firewall-cmd --reload
```
