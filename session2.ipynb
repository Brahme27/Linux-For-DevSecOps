## Computer Architecture, Cloud, Linux, and Server Access

---

## 1 What is a Computer

A computer is an **IP enabled device** that can connect to a network or the internet. Every computer works using four main components. These components are **RAM**, **Storage**, **Operating System**, and **Processor**. All these parts work together to run applications.

---

## 2 Core Components of a Computer

### 2.1 RAM

RAM is **Random Access Memory**. It is used to run active tasks and programs. Whatever we are currently working on is stored in RAM temporarily.

### 2.2 Storage

Storage is used to store data permanently. Files, applications, and operating systems are stored in storage devices.

### 2.3 Operating System (OS)

The Operating System controls hardware and software. It acts as a bridge between the user and the computer hardware.

### 2.4 Processor

The processor is the **brain of the computer**. It executes instructions and performs calculations.

---

## 3 Client Server Architecture

### 3.1 Server

A **server** is a machine used to **host applications**. Applications like Facebook, Google, or Amazon run on servers.

### 3.2 Client

A **client** is software used to access server applications. For example, a **browser** is a client software.

When we open **facebook dot com**, the Facebook application is running on Facebook servers, and our browser acts as a client to access it.

---

### 3.3 Port Forwarding

Port forwarding allows us to **deploy an application on our laptop** and make it accessible over the internet. This helps external users access applications running locally.

---

## 4 Linux and Cloud Basics

### 4.1 Cloud Regions and Availability Zones

A **Region** is a geographical location such as Hyderabad, Mumbai, Singapore, US, or EU.
An **Availability Zone (AZ)** is a physical data center inside a region, such as North Hyderabad and South Hyderabad.
Using **minimum two AZs** ensures **high availability**.

---

### 4.2 Instance and Security

An **Instance** is the same as a **Server** or a **Node**.
A **Firewall** in AWS is called a **Security Group**.
Security groups control network traffic.

* **Inbound rules** control incoming traffic
* **Outbound rules** control outgoing traffic

If we allow **0.0.0.0 slash 0**, it means we are allowing **every computer on the internet** to access the server.

---

## 5 Linux Operating System

### 5.1 Linux Basics

Linux is written using the **C language**.
The **Kernel** is the brain of the operating system.
Linux OS is made of **Kernel plus User Interface**.
Linux is **open source**.

The inventor of Linux is **Linus Torvalds**.

---

### 5.2 Linux Distributions (Flavours)

Linux implementations are called **Distributions or Flavours**.

Some examples are:

* Red Hat Enterprise Linux (RHEL) – Commercial
* CentOS
* Amazon Enterprise Linux
* Fedora
* AlmaLinux
* Ubuntu
* Solaris
* Suse
* Android

Some are **enterprise**, and some are **open source**.

---

### 5.3 AMI and ISO

In AWS, we use **AMI (Amazon Machine Image)**.
AMI is similar to an **ISO file**.
An ISO file creates a complete operating system like **C colon backslash Windows**.
Example AMI is **AWS Linux 2023 AMI**.

Instance types like **t2 micro** and **t3 micro** are commonly used.

## 6 Authentication

Authentication means verifying identity.

### 6.1 Types of Authentication

1 **What you know**
Username and password

2 **What you have**
Username and token or OTP

3 **What you are**
Biometrics such as fingerprints, retina scan, or palm scan

---

## 7 Public Key and Private Key Authentication

Linux servers use **public key and private key authentication**.

### 7.1 Lock and Key Concept

* **Public Key** acts like a **lock**
* **Private Key** acts like a **key**

The server has the public key, and the user has the private key.

The server is accessed using a **public IP address**.

---

### 7.2 Key Generation

Using the command
`ssh-keygen -f <file-name>`
we generate both public and private keys.

* `.pub` file is the **public key**
* `.pem` file is the **private key**

---

## 9 Connecting to AWS Linux Server

### 9.1 Required Details

To connect to a server, we need:

* IP address
* Username
* Private key
* Protocol
* Port

For HTTP, port is **80**.
For Secure Shell (SSH), port is **22**.

---

### 9.2 SSH Access

SSH stands for **Secure Shell**.
Port **22** provides full access to the server.

For **AWS Linux 2023 AMI**, the default username is **ec2-user**.

The connection command is
`ssh -i <private-key> ec2-user@IP`

---
