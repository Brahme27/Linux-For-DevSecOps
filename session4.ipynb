## Vim Editor, User Management, and Linux Administration

---

## 1 Vim Editor

### 1.1 What is Vim Editor

Vim editor is a powerful text editor used in Linux servers to create, view, and edit files. It works mainly using keyboard commands and has different modes.

---

## 2 Vim Editor Modes

Vim editor mainly works in three modes: **ESC mode**, **Command mode**, and **Insert mode**.

---

## 3 ESC Mode (Default Mode)

When we open a file using vim, it opens in **ESC mode by default**. In this mode, we cannot type text, but we can perform operations like copy, paste, delete, and navigation.

### 3.1 ESC Mode Commands

* `u` → undo the last change
* `yy` → copy the current line
* `p` → paste the copied line
* `10p` → paste the copied line 10 times
* `dd` → cut (delete) the current line
* `gg` → move cursor to the top of the file
* `Shift + g` → move cursor to the bottom of the file

---

## 4 Command Mode

To enter **command mode**, press `:` while in ESC mode.
Command mode is used to save, quit, search, replace, and perform file-level operations.

### 4.1 Command Mode Commands

* `:q` → quit the file
* `:wq` → save (write) and quit
* `:q!` → force quit without saving changes
* `:/<word-to-find>` → search the word from top to bottom
* `:?<word-to-find>` → search the word from bottom to top
* `:noh` → remove highlighted search results
* `:set nu` → show line numbers
* `:set nonu` → hide line numbers
* `:28d` → delete line number 28
* `:3s/word-to-find/word-to-replace` → replace first occurrence in line 3
* `:3s/word-to-find/word-to-replace/g` → replace all occurrences in line 3
* `:%s/word-to-find/word-to-replace/g` → replace all occurrences in the entire file
* `:%d` → delete entire file content

---

## 5 Insert Mode

To enter **insert mode**, press `i` from ESC mode.
Insert mode allows us to type and edit text in the file.

---

## 6 Linux Administration

Linux administration involves managing users, groups, permissions, and system configurations.

---

## 7 User Management in Linux

User management is used to create users, assign groups, and manage access.

### 7.1 Creating a User

* `useradd <user-name>`
  This command creates a user and also creates a **primary group with the same name**.

---

### 7.2 Viewing User Information

* `id <user-name>`
  Displays user ID, group ID, and group details.

User information is stored in:

* `/etc/passwd` → contains user information
* `/etc/group` → contains group information

In Linux, a user must have **only one primary group** and **at least one secondary group**.

---

### 7.3 Setting Password

* `passwd <user-name>`
  This command sets or changes the password for a user.

---

## 8 Group Management

### 8.1 Creating a Group

* `groupadd <group-name>`
  This command creates a new group.

---

### 8.2 Adding User to Groups

* `usermod -g devops ramesh`
  Adds user **ramesh** to **devops** as the **primary group**.

* `usermod -aG testing ramesh`
  Adds **testing** group as a **secondary group** to ramesh.

---

## 9 CRUD Concept in User Management

CRUD stands for **Create, Read, Update, and Delete**.

* Create → useradd
* Read → id, passwd file
* Update → usermod
* Delete → userdel

Example:
If an employee is removed from a project, we remove him from the group.
If the employee leaves the company, we delete the user.

* `userdel <user-name>` → deletes the user
* Also remove the user from project groups like devops.

---

## 10 SSH and Authentication

Linux follows **key-based authentication by default** instead of password authentication.

---

## 11 SSH Configuration File

The SSH configuration file is:

`/etc/ssh/sshd_config`

This file contains all SSH related settings.

⚠️ If we make **any mistake in this file**, we may **lose SSH access to the server**.

---

### 11.1 SSH Configuration Commands

* `sshd -t` → checks whether SSH configuration is correct or not
* `systemctl restart sshd` → restarts the SSH service after changes

---

## 12 Key-Based Authentication

In key-based authentication, the user (example **ramesh**) must generate his **public key and private key**.
The public key is stored on the server, and the private key is kept with the user.
This method is more secure than password authentication.
