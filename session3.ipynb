# SESSION 03

## Linux Paths, Users, Commands, and File Operations

---

## 1 Connecting to Linux Server using SSH

To connect to a Linux server in AWS, we use the SSH command.
The basic format of the command is:

`ssh -i <path-to-your-private-key> ec2-user@IP`

Here, the private key is used for authentication, **ec2-user** is the default username for AWS Linux 2023 AMI, and IP is the public IP address of the server.

---

## 2 Absolute Path and Relative Path

### 2.1 Absolute Path

An **absolute path** starts from the root of the server.
It always begins with a forward slash `/`.

Example:
`/c/devops/daws-82s/daws-82s.pem`

This path gives the complete location of the file from the root.

---

### 2.2 Relative Path

A **relative path** starts from the current working directory.

Example:
`daws-82s/daws-82s.pem`

It does not start with `/` and depends on where we are currently located.

---

## 3 Clearing the Terminal

The command **clear** is used to clean the terminal screen so that it becomes easy to read commands again.

---

## 4 Linux Users and Prompt Symbols

In Linux, the terminal prompt symbol tells us which user we are logged in as.

* `$` symbol means **normal user**
* `#` symbol means **root or admin user**

The command `sudo su` is used to switch from a normal user to the root user.

---

## 5 Home Directories in Linux

Each user in Linux has a home directory.

* Normal user home directory
  `/home/<user-name>`

* Root user home directory
  `/root`

If we run `sudo su -`, we directly land inside the root user home directory which is `/root`.

---

## 6 General Linux Command Format

Most Linux commands follow this format:

`<command-name> <options> <inputs>`

Options modify how a command behaves.

* Single dash `-` is used for single character options
* Double dash `--` is used for full word options

---

## 7 Root Directory

The forward slash `/` represents the **root of the server**.
All files and folders in Linux start from this root directory.

---

## 8 Listing Files and Directories

### 8.1 ls Command

The `ls` command is used to list subdirectories and files.

Different variations of `ls` are:

* `ls -l` → long listing format with more details
* `ls -lr` → reverse alphabetical order
* `ls -lt` → newest files on top
* `ls -ltr` → oldest files on top
* `ls -ltrh` → human readable file sizes
* `ls -la` → shows all files including hidden files

---

### 8.2 File Type Indicators

In long listing format, file types are shown as:

* `d` → directory
* `-` → normal file
* `l` → link file

Examples:

* `drwx------` means directory
* `-rw-r--r--` means file
* `lrw-r--r--` means link file

---

## 9 Creating and Viewing Files

### 9.1 Creating Empty Files

The command `touch <file-name>` is used to create an empty file.

---

### 9.2 Writing Content to a File

Using the command:

`cat > devops.txt`
`cat > DevSecOps.txt`

We can enter text into the file.
After typing the content, press **Enter** and then **Ctrl + D** to save.

The symbol `>>` is used to **append content** to an existing file instead of overwriting it.

---

## 10 Creating and Removing Directories

* `mkdir <directory-name>` is used to create a directory
* `rmdir <directory-name>` is used to delete an **empty directory only**

To remove a directory along with its contents, we use:

`rm -r devops`

This command deletes everything inside the directory recursively.

---

## 11 Copying and Navigating Directories

* `cp <source> <destination>` is used to copy files
* `cd ..` is used to move one directory back

---

## 12 Downloading Files using curl and wget

### 12.1 wget Command

The `wget <URL>` command downloads a file from the internet.

---

### 12.2 curl Command

* `curl <URL>` → shows content on the screen
* `curl <URL> -o <path>` → downloads the file with the given name

Example URL:
`https://raw.githubusercontent.com/DAWS-82S/notes/refs/heads/main/session-02.md`

---

## 13 Slash Symbol Meaning

The `/` symbol is used as:

* separator
* delimiter
* fragment identifier

---

## 14 Piping in Linux

Piping allows us to send the output of one command as input to another command.

The pipe symbol is `|`.

Example:
`cat <file-name> | grep <word-to-search>`

This searches for a word inside a file.

---

## 15 grep Command

The `grep` command is used to search for specific words in files.

Syntax:
`grep <word-to-search> <file>`

---

## 16 cut Command

The `cut` command is used to extract specific parts of text.

Example:

`echo "https://raw.githubusercontent.com/DAWS-82S/notes/refs/heads/main/session-02.md" | cut -d "/" -f9`

Output:
`session-02.md`

Here, `-d "/"` means delimiter is slash and `-f9` means field number 9.

---

## 17 awk Command

The `awk` command is used for advanced text processing.

Example:

`echo "https://raw.githubusercontent.com/DAWS-82S/notes/refs/heads/main/session-02.md" | awk -F "/" '{print $NF}'`

This prints the last field of the text.

---

## 18 Viewing Users in Linux Server

All users in Linux are stored in the **passwd** file.

Command:

`awk -F ":" '{print $1}' passwd`

This prints all usernames.

---

## 19 head and tail Commands

* `head passwd` → shows first 10 lines
* `tail -n 4 passwd` → shows last 4 lines

Combination example:

`head -n 10 passwd | tail -n 7`

This prints lines from **4 to 10**.

---

## 20 find Command

The `find` command is used to search files in the system.

Syntax:
`find <location> -name "<file-name>"`

Example:
`find / -name "passwd"`

This searches for the passwd file from the root directory.

---

The **vim editor** is a powerful text editor used to create and edit files in Linux servers.
