## 1 Linux Permissions

In Linux, permissions decide **who can read, write, or execute** a file or folder. Permissions are represented using numbers.

* **Read (R)** = 4
* **Write (W)** = 2
* **Execute (X)** = 1

---

## 2 Permission Structure of a File

A file permission looks like this:

`- rw- r-- r-- ec2-user ec2-user`

This is divided into three parts.

### 2.1 Owner Permissions

The **owner** is the user who created the file.
In this example, **ec2-user** is the owner and has **read and write** permissions.

### 2.2 Group Permissions

The **group** also has **read-only** permission.

### 2.3 Others Permissions

**Others** means anyone who is not the owner and not part of the group.
Others also have **read-only** permission.

---

## 3 Who Can Change Permissions

Only the **file owner** or the **root user** can change permissions of a file or folder.

---

## 4 Permission Symbols

Linux uses symbols to represent users:

* **u** → owner or user
* **g** → group
* **o** → others

---

## 5 Changing Permissions Using chmod

### 5.1 Symbolic Mode

* `chmod o+w devsecops.txt`
  Adds write permission to others

* `chmod o-r suresh.txt`
  Removes read permission from others

* `chmod ugo+rwx suresh.txt`
  Gives read, write, and execute permission to everyone

---

### 5.2 Numeric Mode

* `chmod 740 suresh.txt`
  Owner → read write execute
  Group → read only
  Others → no permission

* `chmod 777 suresh.txt`
  Everyone gets full permission

* `chmod 755 suresh.txt`
  Owner → full permission
  Group and others → read and execute

---

## 6 SSH Key Setup for a User

If a user named **suresh** needs server access, the admin should ask for **suresh public key**.

Steps followed by root user:

1 Create a `.ssh` folder inside `/home/suresh`
2 Change ownership of `.ssh` folder to suresh

`chown -R suresh:suresh .ssh`

General format:
`chown <user>:<group> file or folder`

---

## 7 authorized_keys File

Inside the `.ssh` folder, we create a file named **authorized_keys**.
This file contains the public key of the user.

* Owner of this file must be **suresh**
* Maximum permission allowed is **600**

The file path will be:
`/home/suresh/.ssh/authorized_keys`

---

## 8 SSH Login as User

Once setup is complete, the user can connect using:

`ssh -i suresh.pem suresh@IP`

## 9 Package Management

Package management is used to install, remove, and manage software.

In Windows, laptops are configured with URLs to pull updates.
In Linux, packages are downloaded from repositories.

Usually, **one package depends on other packages**.

---

### 9.1 Installing Packages

* `yum install <package-name>`
* `dnf install <package-name>` → used in Amazon Linux
* `apt-get install <package-name>` → used in Ubuntu

---

### 9.2 Removing Packages

* `dnf remove <package-name>`

---

### 9.3 Listing Packages

* `dnf list installed` → shows already installed packages
* `dnf list available` → shows all available packages

---

## 10 Service Management

Some packages run as **services**.

Example: SSH, Nginx, Apache.

To connect as user:
`ssh -i <private-key> suresh@IP`

The request goes to the server IP and checks whether **SSH is running on port 22**.

---

### 10.1 Checking Service Status

* `systemctl status sshd`

---

### 10.2 HTTP Service Example (Nginx)

Install nginx:
`dnf install nginx -y`

Start nginx service:
`systemctl start nginx`

Access from browser:
`http://IP:80`

---

### 10.3 Managing Services

* `systemctl stop nginx` → stops service
* `systemctl restart nginx` → restarts service
* `systemctl enable nginx` → starts service automatically on reboot

Running `systemctl start git` is invalid because **git is a utility, not a service**.

Some packages are **command-line utilities**, and some are **service-based packages**.

---

## 11 Process Management

In Linux, **everything is a process**.

To understand parent-child process relationship, consider this example:

* Team Lead gives TASK-1 to Senior
* Senior gives TASK-2 to Junior
* Junior gives TASK-3 to Fresher

Here:

* TASK-2 is parent of TASK-3
* TASK-1 is parent of TASK-2

---

### 11.1 Process Example

When we run:
`echo "Hello World"`

Linux creates a **process instance ID**, executes the command, prints output, and then marks the process as completed.

---

### 11.2 PPID

PPID means **Parent Process ID**.
Every process has a parent process except the first system process.

---

### 11.3 Foreground and Background

Processes can run in:

* Foreground
* Background

---

### 11.4 Killing a Process

* `kill PID` → request to stop the process
* `kill -9 PID` → force stop the process

---

## 12 Network Management

Network management is used to check ports, services, and connections.

### 12.1 Network Commands

* `netstat -lntp` → checks which ports are open and listening
* `systemctl status nginx` → checks nginx service status
* `ps -ef | grep nginx` → checks nginx process
* `netstat -lntp` → confirms whether port is open or not
